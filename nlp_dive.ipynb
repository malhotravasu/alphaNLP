```python
sent = 'Hi, I am Vasu Malhotra from DTU.' #Demo Sentence
```

```python
from nltk import word_tokenize, pos_tag, ne_chunk
```

```python
nltk_token = word_tokenize(sent) #Words tokenize
nltk_token
```

```python
nltk_tag = pos_tag(nltk_token) #Part of speech tagging
nltk_tag
```

```python
nltk_chunk = ne_chunk(nltk_tag) #NER Named Entity Recognition
nltk_chunk
```

```python
nltk_tree = ne_chunk(pos_tag(word_tokenize(sent))) #All three combined
nltk_tree
```

```python
print(nltk_tree.__repr__())
```
